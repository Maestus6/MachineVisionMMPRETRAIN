```python
import os
import json
import shutil
from google.colab import drive
```

```python
drive.mount('/content/gdrive')

def create_folders(mainFolder, subfolders): #creating folder for each directories
    for categoryFolder in subfolders:
        os.makedirs(os.path.join(mainFolder, categoryFolder), exist_ok=True)
```

```python
#moving images into folders,
#for loop inside of for loop isnt ideal, will fix it if i have time later
def divideImages(data, mainFolder, categoryFolder):
    for category, images in data.items():
        category_folder = os.path.join(categoryFolder, category)
        os.makedirs(category_folder, exist_ok=True)
        for image in images:
            source_path = os.path.join(mainFolder, image)
            if os.path.exists(source_path):
                shutil.move(source_path, category_folder)
```

```python
def main():
    mainFolder = "/content/gdrive/My Drive/picsandjsons"
    trainingSetFolder = os.path.join(mainFolder, "trainingSet")
    validationSetFolder = os.path.join(mainFolder, "validationSet")

    #creating the set data
    create_folders(mainFolder, ["trainingSet", "validationSet"])

    #loading the jsons
    with open("/content/gdrive/My Drive/picsandjsons/train_set.json", 'r') as file:
        trainingData = json.load(file)
    with open("/content/gdrive/My Drive/picsandjsons/val_set.json", 'r') as file:
        validationData = json.load(file)

    #dividing the images between folders
    divideImages(trainingData, mainFolder, trainingSetFolder)
    divideImages(validationData, mainFolder, validationSetFolder)
```

```python
#calling the main function
if __name__ == "__main__":
    main()
```
